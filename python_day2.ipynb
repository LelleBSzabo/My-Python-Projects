```python
password = "password"
user_entry = input("Enter the password:")

if user_entry == password:
  print("Password is correct. Login successful.")

while user_entry != password:
  print("Password is incorrect. Login Failed")
  user_entry = input("Try again:")
  if user_entry == password:
    print("Password is correct. Login successful.")
    break
```

```python
secret_number = 8
tries_left = 3

while tries_left > 0:
    guess = int(input("Guess the secret number (1-10): "))

    if guess == secret_number:
        print("🎉 Correct! You win!")
        break
    elif guess < secret_number:
        print("Too low.")
    else:
        print("Too high.")

    tries_left -= 1
    print(f"Tries left: {tries_left}")

if tries_left == 0:
    print("😢 Game over. The correct number was", secret_number)
```



Go through the integers from 1 to 100.

If a number is divisible by 3, print "fizz."

If a number is divisible by 5, print "buzz."

If a number is both divisible by 3 and by 5, print "fizzbuzz."

Otherwise, print just the number.

Use loops and if statements.  

```python
n = 1

while n <= 100:
    if n % 3 == 0 and n % 5 == 0:
        print("fizzbuzz")
    elif n % 3 == 0:
        print("fizz")
    elif n % 5 == 0:
        print("buzz")
    else:
        print(n)

    n += 1
```
